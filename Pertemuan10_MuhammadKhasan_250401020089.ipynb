```python
# Langkah 1: Muat dan Eksplorasi Data
import pandas as pd

df = pd.read_csv("telco_churn.csv")
print("\nUkuran dataset:")
print(df.shape)

print("\nInformasi dataset:")
print(df.info())

print("\nProporsi kelas Churn:")
print(df["Churn"].value_counts(normalize=True))
```

```python
# Langkah 2: Preprocessing
from sklearn.model_selection import train_test_split

## TODO: encoding fitur kategorikal (mis. pd.get_dummies)
df = pd.get_dummies(df, drop_first=True)

## TODO: pisahkan X (fitur) dan y (target = Churn)
X = df.drop('Churn_Yes',axis=1)
y = df['Churn_Yes']

## Pembagian data (train-test split)
X_tr, X_te, y_tr, y_te = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
## Ukuran data (train-test split)
print("Ukuran X_train :", X_tr.shape)
print("Ukuran X_test  :", X_te.shape)
print("Ukuran y_train :", y_tr.shape)
print("Ukuran y_test  :", y_te.shape)
```

```python
# Langkah 3: Latih Model
from sklearn.ensemble import RandomForestClassifier

## Membangun Model Random Forest Classifier
rf = RandomForestClassifier(
    n_estimators=300,
    class_weight="balanced",
    random_state=42
)

rf.fit(X_tr, y_tr)
```

```python
# Langkah 4: Evaluasi
from sklearn.metrics import classification_report, roc_auc_score

## Menghitung Prediksi kelas
y_pred = rf.predict(X_te)

## Menghitung Probabilitas kelas 1 (Churn)
y_prob = rf.predict_proba(X_te)[:, 1]

## Menampilkan Precision, Recall, F1-score
print(classification_report(y_te, y_pred))

## Menghitung dan Menampilkan ROC-AUC
roc = roc_auc_score(y_te, y_prob)
print("ROC-AUC:", roc)
```

```python
# Langkah 5: Prediksi Probabilitas dan Simpulkan
# TODO: hitung probabilitas churn (predict_proba)
y_prob = rf.predict_proba(X_te)[:, 1]

# Membuat tabel hasil prediksi
hasil_prediksi = pd.DataFrame({
    "No Data": range(1, len(y_prob)+1),
    "Probabilitas_Churn": y_prob,
    "Persentase_Churn (%)": y_prob * 100
})

# TODO: Menampilkan 10 probabilitas churn pertama
print(hasil_prediksi.head(10))
```

**KESIMPULAN :**

Model Random Forest berhasil menghasilkan probabilitas churn untuk setiap pelanggan menggunakan metode predict_proba. Berdasarkan hasil prediksi, model mampu memberikan estimasi tingkat kemungkinan pelanggan melakukan churn berdasarkan nilai probabilitas yang dihasilkan. Nilai probabilitas yang semakin mendekati 1 (100%) menunjukkan bahwa pelanggan memiliki kemungkinan lebih tinggi untuk churn, sedangkan nilai probabilitas yang semakin mendekati 0 (0%) menunjukkan bahwa pelanggan memiliki kemungkinan lebih besar untuk tetap menggunakan layanan.

Hasil prediksi probabilitas churn ini dapat dimanfaatkan untuk mengidentifikasi pelanggan yang memiliki potensi risiko churn lebih tinggi maupun pelanggan yang cenderung bertahan. Informasi tersebut dapat digunakan sebagai dasar dalam menyusun strategi retensi pelanggan yang lebih tepat sasaran, sehingga dapat dilakukan tindakan pencegahan untuk mempertahankan pelanggan dan mengurangi potensi kehilangan pelanggan.
