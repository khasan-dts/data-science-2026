```python
# Memuat Dataset Titanic
import pandas as pd
import numpy as np

URL = "https://raw.github.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)

df
```

```python
# Q1 : Jumlah baris dan kolom dataset Titanic
print(df.shape, df.columns.tolist())
```

Dataset Titanic memiliki 891 baris dan 12 kolom yang terdiri dari : PassengerId, Survived, Pclass, Name, Sex, Age, SibSp, Parch, Ticket, Fare, Cabin, Embarked

```python
# Q2 : kolom mana yang memiliki nilai kosong (missing) pada dataset Titanic? berapa dan proprosinya
print(df.isnull().sum())
print(df.isnull().mean())
```

Dataset Titanic memiliki 3 kolom kosong (missing), Yaitu:

1. **Age** dengan jumlah 177 dan proporsi 0,19%
2. **Cabin** dengan jumlah 687 dan proporsi 0,77%
3. **Embarked** dengan jumlah 2 dan proporsi 0,02%





```python
# Q3 : Berapa persen penumpang yang selamat (Survived=1)
selamat = df[df['Survived']==1]
persentase = len(selamat)/len(df)

print ("Total Penumpang :", len(df))
print("Penumpang selamat :", df['Survived'].sum())
print("Persentase selamat (%) :", persentase * 100)
```

Persentase penumpang selamat sejumlah **38.38%** dari total penumpang

```python
# Q4: Rata-rata usia penumpang wanita kelas 1 yang selamat
wk1_selamat = df[(df['Sex']=='female') &
                 (df['Pclass']==1) &
                 (df['Survived']==1)]
print ("Rata-Rata Usia Wanita Kelas I selamat :", wk1_selamat['Age'].mean())
```

Berdasarkan Dataset Titanic Rata-rata Usia Penumpang Kelas I yang selamat adalah 35 Tahun

```python
# Q5: Kelas penumpang yang memiliki tingkat keselamatan tertinggi
hasil = df.groupby('Pclass').agg(
        total=('PassengerId', 'count'),
        selamat=('Survived','sum'),
        persen = ('Survived','mean'),
).round(3)
print(hasil)
```

Tingkat keselamatan penumpang tertinggi terdapat pada kelas I yaitu **sejumlah** **136 penumpang** atau 0,630%
