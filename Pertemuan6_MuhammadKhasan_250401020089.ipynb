```python
# --------------------------------------------------
# LANGKAH 1 - Load & EDA Singkat
# --------------------------------------------------

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('titanic')

# Pilih kolom
cols = ['pclass', 'sex', 'age', 'sibsp',
        'parch', 'fare', 'embarked', 'survived']
df = df[cols].copy()

# Informasi ukuran dataset
print("Shape:", df.shape)

# Missing values
print("\nMissing values:")
print(df.isnull().sum())

# Distribusi target
print("\nDistribusi target:")
print(df['survived'].value_counts(normalize=True).round(3))
```

```python
# --------------------------------------------------
# LANGKAH 2 - Handling Missing Values
# --------------------------------------------------

# Age: isi missing value dengan median
df['age'] = df['age'].fillna(df['age'].median())

# Embarked: isi missing value dengan modus
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

# Cek missing value setelah handling
print('Missing setelah handling:')
print(df.isnull().sum())  # Semua harus 0
```

```python
# --------------------------------------------------
# LANGKAH 3 - Encoding Kategorikal
# --------------------------------------------------

# One-Hot Encoding untuk kolom kategorikal
df = pd.get_dummies(
    df,
    columns=['sex', 'embarked'],
    drop_first=True,   # hindari dummy variable trap
    dtype=int
)

# Tampilkan nama kolom setelah encoding
print('Kolom setelah encoding:')
print(df.columns.tolist())

# Hasil yang diharapkan:
# ['pclass', 'age', 'sibsp', 'parch', 'fare', 'survived',
#  'sex_male', 'embarked_Q', 'embarked_S']
```

```python
# --------------------------------------------------
# LANGKAH 4 - Train-Test Split
# --------------------------------------------------

from sklearn.model_selection import train_test_split

# Pisahkan fitur (X) dan target (y)
X = df.drop('survived', axis=1)
y = df['survived']

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,      # 20% data test
    random_state=42,    # agar hasil selalu sama
    stratify=y          # menjaga proporsi kelas
)

# Jumlah data train dan test
print(f'Train: {X_train.shape[0]} baris')
print(f'Test : {X_test.shape[0]} baris')

# Cek proporsi kelas
print('\nProporsi survived di Train:')
print(y_train.value_counts(normalize=True).round(3))

print('\nProporsi survived di Test:')
print(y_test.value_counts(normalize=True).round(3))
```

```python
# --------------------------------------------------
# LANGKAH 5 - Feature Scaling
# --------------------------------------------------

from sklearn.preprocessing import StandardScaler

# Kolom numerik yang akan di-scale
num_cols = ['pclass', 'age', 'sibsp', 'parch', 'fare']

# Membuat objek scaler
scaler = StandardScaler()

# Fit + transform data training
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])

# Transform data testing
X_test[num_cols] = scaler.transform(X_test[num_cols])

# Menampilkan mean dan standar deviasi scaler
print('Mean scaler (dari train):', scaler.mean_.round(2))
print('Std scaler (dari train):', scaler.scale_.round(2))
print()

# Contoh hasil scaling
print('Contoh X_train setelah scaling:')
print(X_train.head(3).round(3))
print('\nData siap dilatih model Machine Learning!')

# Ukuran dataset
print(f'X_train: {X_train.shape}, y_train: {y_train.shape}')
print(f'X_test : {X_test.shape}, y_test : {y_test.shape}')
```
